# 1. 创建数据 -- 获取数据

**1.1 生成测试数据的函数**

```python
# 设置种子
np.seed(111)

# 生成测试数据的函数
def CreateDataSet(Number=1):

    Output = [] 

    for i in range(Number):

        # 创建一个按周计算的日期范围(每个周一开始)
        rng = pd.date_range(start='1/1/2009', end='12/31/2012', freq='W-MON')

        # 创建一些随机数
        data = np.randint(low=25, high=1000, size=len(rng))

        # 状态池
        status = [1, 2, 3]

        # 创建一个随机的状态列表
        random_status = [status[np.randint(low=0, high=len(status))] 
                         for i in range(len(rng))]

        # 行政州(state)的列表
        states = ['GA','FL','fl','NY','NJ','TX']

        # 创建一个行政周的随机列表
        random_states = [states[np.randint(low=0, high=len(states))] 
                         for i in range(len(rng))]

        Output.extend(zip(random_states, random_status, data, rng))

    return Output
```

**1.2 生成测试数据并放入df**

```python
dataset = CreateDataSet(4)
df = pd.DataFrame(data=dataset, 
                  columns=['State','Status','CustomerCount','StatusDate'])
```

**2.5 用groupyby()和sum()这两个函数对数据进行压缩处理**

我们的任务是创建一个新的 dataframe，然后对数据进行压缩处理，使得每一个 State 和 StatusDate 组合代表一天的客户数量。我们可以忽略 Status 列因为这一列我们之前处理过只有 1 这个值了。     
    
要完成这个操作，我们使用 dataframe 的 groupyby() 和 sum() 这两个函数。    
    
注意，我们要使用 reset_index。    
如果我们不这么做，我们将无法同时用 State 和 StatusDate 这两列来做分组，因为 groupby 函数需要列(columns)来做为输入(译者注: StatusDate 目前是 index，不是 column)。   
reset_index 函数将把 dataframen 中作为索引(index)的 StatusDate 变回普通的列。

```python
# 先 reset_index，然后按照State和 StatusDate来做分组(groupby)
Daily = df.reset_index().groupby(['State','StatusDate']).sum()
Daily.head()
```

**2.9 将特定范围之外的数据移除**

我们假设每个月的客户数量应该是保持相对稳定的。 在一个月内任何在这个特定范围之外的数据都可以从数据集中移除。         
最终的结果应该更加的平滑并且图形不会有尖刺。      
     
Daily.index.get_level_values(0)       表示state         
Daily.index.get_level_values(1).year    表示StatusDate的年份               
Daily.index.get_level_values(1).month   表示StatusDate的月份     
        
看过前面的图形我们意识到这些图形不是高斯分布的(gaussian distribution)，  
这意味着我们不能使用均值(mean)和标准差(stDev)这些统计量。 我们将使用百分位数(percentile)。    
请注意这里也会有把好数据消除掉的风险。     
DataFrame.quantile(q) 是四分卫函数,就是均等分为四份. pandas中q取0.25 0.5 0.75 .             
x.quantile(q=.25) - (1.5*x.quantile(q=.75)-x.quantile(q=.25)) )    
             
这里用到 transform 而不是 apply。   
原因是,transform 将会保持dataframe矩阵的形状(shape)(就是行列数不变)而apply会改变矩阵的形状。




```python
# 移除离群值


# 先按照state ,StatusDate的年份,StatusDate的月份进行分组
StateYearMonth = Daily.groupby([Daily.index.get_level_values(0), 
                                Daily.index.get_level_values(1).year, 
                                Daily.index.get_level_values(1).month])

#确定范围
Daily['Lower'] = StateYearMonth['CustomerCount'].transform( 
    lambda x: x.quantile(q=.25) - (1.5*x.quantile(q=.75)-x.quantile(q=.25)) )

Daily['Upper'] = StateYearMonth['CustomerCount'].transform( 
    lambda x: x.quantile(q=.75) + (1.5*x.quantile(q=.75)-x.quantile(q=.25)) )

# 新增了Outlier列,计算离群值 .
# 当Daily['CustomerCount'] 不在[Daily['Lower'] , Daily['Upper']]范围内, 
# 即 Outlier 列如果是Ture, 则为离群值.
Daily['Outlier']=(Daily['CustomerCount']<Daily['Lower'])|(Daily['CustomerCount']>Daily['Upper']) 

# 移除离群值
Daily = Daily[Daily['Outlier'] == False]

Daily.head()
```

**2.10 分组并求每组最大值**

我们创建一个单独的 dataframe，叫 ALL，仅用 StatusDate 来为 Daily 数据集做索引。    
我们简单地去掉 State 这一列。 Max 列则代表了每一个月最大的客户数量。 Max 列是用来是的图形更顺滑的。

```python
# 合并所有市场的

# 按日期计算出最大的客户数
ALL = pd.DataFrame(Daily['CustomerCount'].groupby(Daily.index.get_level_values(1)).sum())
ALL.columns = ['CustomerCount'] # rename column

# 按照年和月来分组
YearMonth = ALL.groupby([lambda x: x.year, lambda x: x.month])

# 增加一列Max, 找出每一个年和月的组合中最大的客户数
ALL['Max'] = YearMonth['CustomerCount'].transform(lambda x: x.max())
ALL.head()
```

从上面的 ALL dataframe 中可以看到，在 2009年1月的这个月份，最大的客户数是 901。   
如果我们使用 apply 的方式，我们将会得到一个以 (年 和 月）组合作为索引的 dataframe ，只有 Max 这一列有901这个值。

**2.11 可视化的展示当前的客户数是否达到目标值**

如果当前的客户数达到了公司制定的一定的目标值，这也会是一个很有趣的度量值。     
现在的任务是可视化的展示当前的客户数是否达到了下面列出的目标值。      
我们把这些目标叫做 BHAG (Big Hairy Annual Goal，年度战略目标)。     
12/31/2011 - 1,000 客户        
12/31/2012 - 2,000 客户                
12/31/2013 - 3,000 客户        

用 date_range 函数来创建日期。    
定义:    
date_range(start=None, end=None, periods=None, freq='D', tz=None, normalize=False, name=None, closed=None)   
文档(Docstring): 返回一个固定频率的日期时间索引，用日历天作为默认的频率       
把频率设定为 A 或者是“年度” 我们将等到上述三个年份。       

```python
# 创建 BHAG 数据
data = [1000, 2000, 3000]
idx = pd.date_range(start='12/31/2011', end='12/31/2013', freq='A')
BHAG = pd.DataFrame(data, index=idx, columns=['BHAG'])
BHAG
```

我们之前学过用 concat 把 dataframe 合并。     
记住，当我们设置 axis = 0 时我们按列合并     
(即直接把列合并进去，行方向数据缺失用NaN来填充）。   

```python
# 把 BHAG 和 ALL 两个数据集合并在一起
combined = pd.concat([ALL,BHAG], axis=0)
combined = combined.sort_index(axis=0)
combined.tail()
```

```python
fig, axes = plt.subplots(figsize=(12, 7))

combined['BHAG'].fillna(method='pad').plot(color='green', label='BHAG')
combined['Max'].plot(color='blue', label='All Markets')
plt.legend(loc='best');
```

**2.12 预测下一个年度的客户数**

```python
# 把已经合并的 dataframe 按照 Year 来分组，并且计算出年度的最大客户数。 
# 这样每一行就是一个年度的数据。

Year = combined.groupby(lambda x: x.year).max()
Year
```

```python
# 增加一列，表示为每一年比上一年变化的百分比
Year['YR_PCT_Change'] = Year['Max'].pct_change(periods=1)
Year
```

要得到下一个年度末的客户数，我们假定当前的增长速率是维持恒定的。      
我们按照这个增长速率来预测下一个年度的客户数量。    

```python
(1 + Year.ix['2012','YR_PCT_Change']) * Year.ix['2012,'Max']
```
