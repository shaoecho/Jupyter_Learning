```python

import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(10,8))     # 建立一个大小为10*8的画板
ax1 = fig.add_subplot(331)           # 在画板上添加3*3个画布，位置是第1个
ax2 = fig.add_subplot(3,3,2)
ax3 = fig.add_subplot(3,3,3)
ax4 = fig.add_subplot(334)
ax5 = fig.add_subplot(3,3,5)
ax6 = fig.add_subplot(3,3,6)
ax7 = fig.add_subplot(3,3,7)
ax8 = fig.add_subplot(3,3,8)
ax9 = fig.add_subplot(3,3,9)

# 图1: 折线图  plot
ax1.plot(np.random.randn(10))  

# 图2: 散点图  scatter 
ax2.scatter(np.random.randn(10),np.arange(10),color='r')  

# 图3: 柱形图  hist
ax3.hist(np.random.randn(20),bins=10,alpha=0.3) 

# 图4: 直方图  bar
ax4.bar(np.arange(10),np.random.randn(10)) 

# 图5: 饼形图 pie
ax5.pie(np.random.randint(1,15,5),explode=[0,0,0.2,0,0])  

# 图6: 直线图和直方图结合
x = np.arange(10)
y = np.random.randn(10)
ax6.plot(x,y,color='green')
ax6.bar(x,y,color='k')

#plt.tight_layout() #避免出现叠影
#plt.show()
```

```python
# 图7: 针对DataFrame作散点图 ????

# 生成数据,np.random.randn(100,4)表示生成维度为100行n列的矩阵,且数据满足标准正太分布 
data = DataFrame(np.random.randn(1000,4),
                 columns=['one','two','three','four'])   
data

# 作图
data.plot(x='one',y='two',kind='scatter',ax=ax7)  
plt.show() 
```

```python
# 图8: 针对DataFrame的做直方图  ????
data2 = DataFrame(np.random.randint(0,20,(10,2)),columns=['a','b'])


data2.plot(x='a',y='b',kind='bar',ax=ax8,color='red',legend=False)
data2.plot(x='a',y='b',kind='barh',color='m',ax=ax9)
```
